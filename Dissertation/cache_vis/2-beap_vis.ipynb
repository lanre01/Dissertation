```python
# use to visuallise cache

import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("../bench_results_cache/2-nbeap.csv")
df.head()
```

```python
# visualise 2-beap 
# visuallise push


otherCols = ["op", "size"]

colsToPlot = [col  for col in df.columns if not col in otherCols]
df_pushRandom = df[df["op"] == "PushRandom"] 
df_pushRandom = df_pushRandom.sort_values("size")

plt.figure(figsize=(10,6))
plt.title(f"Plot of 2-beap push random performance overtime")
plt.xlabel("Input Size")
plt.ylabel("time in ns")
plt.xscale("log", base=2)
plt.yscale("log", base=2)
# plt.ylim(bottom=0)

x = df_pushRandom["size"].to_numpy()

for col in colsToPlot:
    y = df_pushRandom[col].to_numpy()
    plt.plot(x, y, marker="o", label=col)
    #print(y)

plt.grid(True, which="both", linestyle=":")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# visualise 2-beap 
# visuallise extract
df = pd.read_csv("../bench_results_cache/2-nbeap.csv")
df.head()

otherCols = ["op", "size"]

colsToPlot = [col  for col in df.columns if not col in otherCols]
df_pushRandom = df[df["op"] == "Extract"] 
df_pushRandom = df_pushRandom.sort_values("size")

plt.figure(figsize=(10,6))
plt.title(f"Plot of 2-beap extract performance overtime")
plt.xlabel("Input Size")
plt.ylabel("time in ns")
plt.xscale("log", base=2)
plt.yscale("log", base=2)
# plt.ylim(bottom=0)

x = df_pushRandom["size"].to_numpy()

for col in colsToPlot:
    y = df_pushRandom[col].to_numpy()
    plt.plot(x, y, marker="o", label=col)
    print(y)

plt.grid(True, which="both", linestyle=":")
plt.legend()
plt.tight_layout()
plt.show()
```
