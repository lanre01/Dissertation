```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("../bench_results_cache/3-nbeap.csv")
df.head()
```

```python
otherCols = ["op", "size"]

labelMap = {
    "cpu_time_7withoutBubbleUp" : "withoutBubbleUp", 
    "cpu_time_8withBubbleUp" : "withBubbleUp"
            }
colsToPlot = ["cpu_time_7withoutBubbleUp", "cpu_time_8withBubbleUp"]
df_extract = df[df["op"] == "Extract"] 
df_extract = df_extract.sort_values("size")

plt.figure(figsize=(10,6))
plt.title(f"Performance Comparison of Extraction Methods in an 8-Beap")
plt.xlabel("Input Size")
plt.ylabel("time in ns")
# plt.xscale("log", base=2)
# plt.yscale("log", base=2)
# plt.ylim(bottom=0)

x = df_extract["size"].to_numpy()

for col in colsToPlot:
    y = df_extract[col].to_numpy()
    plt.plot(x, y, marker="o", label=labelMap[col])
    #print(y)

plt.grid(True, which="both", linestyle=":")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("../bench_results_cache/3-nbeap.csv")
df.head()
```

```python


labelMap = {
    "cpu_time1withoutsqrOpt": "withoutsqrt",
    "cpu_time_3withOpt": "withsqrt"
}

baseCols = ["cpu_time1withoutsqrOpt", "cpu_time_3withOpt"]

df_extract = df[df["op"] == "Extract"].copy()
df_extract["size"] = pd.to_numeric(df_extract["size"])

df_plot = df_extract.pivot(index="size", columns="stat", values=baseCols)
df_plot = df_plot.sort_index()

plt.figure(figsize=(10, 6))
plt.title("Performance Comparison of Extraction Methods in a 3-Beap")
plt.xlabel("Input Size")
plt.ylabel("Time (ns)")
plt.xscale("log", base=2)
plt.yscale("log", base=2)

x = df_plot.index.to_numpy()

for base in baseCols:
    y_mean = df_plot[(base, "median")].to_numpy()
    plt.plot(x, y_mean, marker="o", label=f"{labelMap[base]} (median)")

plt.grid(True, which="both", linestyle=":")
plt.legend()
plt.tight_layout()
plt.show()
```
